```python
# Movies ratings 
rating_9 = 0
rating_8 = 0
rating_less = 0
for num in movies.values:
    if num[5] >= 9:
        rating_9 += 1
    elif num[5] <= 9 and num[5] >= 8:
        rating_8 += 1
    else:
        rating_less += 1

print('The rating is higher than 9:', rating_9)
print('The rating is between 9 and 8:', rating_8)
print('The ratings is less than 8:', rating_less)
```

```python
# A bar graph
labels = ['>= 9', '9 between 8', '<= 8']
values = [rating_9, rating_8, rating_less]
plt.ylabel('Number of Movies')
plt.xlabel('Rating')
plt.title('Movies ratings')
plt.bar(labels, values)
plt.figure(figsize=(6, 4))
plt.show()
```

```python
# TOP-5 movies with the highest gross
movies['Gross'] = movies['Gross'].astype(str).apply(lambda x: int(float("".join(x.split(",")))))
movies1 = movies.sort_values(by = 'Gross', ascending = False) 
movies1.head(5)
```
